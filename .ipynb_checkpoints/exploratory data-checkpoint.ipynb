```python
world_population_df = pops_df
```

```python
# Calculate growth percentage from 1970 to 2022
# Handling cases with missing data by excluding them from calculations
world_population_df['Growth 1970-2022 (%)'] = ((world_population_df['2022 Population'] - world_population_df['1970 Population']) / 
                                               world_population_df['1970 Population']) * 100

# Rank countries based on the calculated growth percentage, sorting in descending order
growth_rank_df = world_population_df[['Country', '1970 Population', '2022 Population', 'Growth 1970-2022 (%)']].dropna()
growth_rank_df = growth_rank_df.sort_values(by='Growth 1970-2022 (%)', ascending=False).reset_index(drop=True)

# Display the top 10 countries with the greatest growth
growth_rank_df.head(10)
```

```python
# Filter for countries with a negative growth percentage from 1970 to 2022
population_decrease_df = growth_rank_df[growth_rank_df['Growth 1970-2022 (%)'] < 0]

# Display countries with an overall decrease in population
population_decrease_df.reset_index(drop=True)
```

```python
#Check japan
japan_population_data = pops_df


japan_population_data = growth_rank_df[growth_rank_df['Country'] == 'Japan']
japan_population_data
```

```python
#Check between 2000 and 2023
# Calculate the growth percentage from 2000 to 2022
world_population_df['Growth 2000-2022 (%)'] = (
    (world_population_df['2022 Population'] - world_population_df['2000 Population']) / 
    world_population_df['2000 Population']
) * 100

# Filter for countries with a negative growth percentage from 2000 to 2022
population_decrease_2000_2022 = world_population_df[
    world_population_df['Growth 2000-2022 (%)'] < 0
]

# Display countries with an overall decrease in population from 2000 to 2022
population_decrease_2000_2022[['Country', '2000 Population', '2022 Population', 'Growth 2000-2022 (%)']]

```

```python
#population_decrease_2000_2022['Country']
countries = population_decrease_2000_2022['Country'].tolist()
print(countries)
```

```python
#graph testing
# Bar chart for percentage decline comparison
countries_decline = population_decrease_df[['Country', 'Growth 1970-2022 (%)']]
countries_decline.set_index('Country').plot(kind='bar', color='tomato', legend=False)
plt.title("Population Decline Percentage by Country (1970 - 2022)")
plt.xlabel("Country")
plt.ylabel("Population Decline (%)")
plt.show()
```

```python
# Filter for countries with population decline
countries_decline = ['Ukraine', 'Latvia', 'Lithuania', 'Bulgaria']  # Example countries with decline
decline_data = world_population_df[world_population_df['Country'].isin(countries_decline)]

# Plot population over years for each country
years = ['1970 Population', '1980 Population', '1990 Population', '2000 Population', '2010 Population', '2022 Population']
for country in countries_decline:
    plt.plot(years, decline_data[decline_data['Country'] == country][years].values.flatten(), label=country)

plt.xlabel('Year')
plt.ylabel('Population')
plt.title('Population Decline Over Time for Selected Countries')
plt.legend()
plt.show()
```

```python

# Load the dataset
file_path = Path("Resources/world_population.csv")
df = pd.read_csv(file_path)

# Melt the DataFrame to long format for better plotting
df_melted = df.melt(id_vars=['Country', 'Continent'], 
                    value_vars=['1970 Population', '1980 Population', '1990 Population', '2000 Population', '2010 Population', '2015 Population', '2020 Population', '2022 Population'],
                    var_name='Year', value_name='Population')

# Clean the 'Year' column to keep only the year number
df_melted['Year'] = df_melted['Year'].str.extract('(\d{4})').astype(int)

# Plotting the population of all countries over time
plt.figure(figsize=(15, 10))

# Plot each country
for country in df['Country'].unique():
    country_data = df_melted[df_melted['Country'] == country]
    plt.plot(country_data['Year'], country_data['Population'], label=country)

plt.title('Population of All Countries Over Time')
plt.xlabel('Year')
plt.ylabel('Population')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', fontsize='small', ncol=2)
plt.show()
```

```python
# Find the top 5 countries with the highest population in 2022
top_5_countries_2022 = df.nlargest(5, '2022 Population')

# Display the top 5 countries
print(top_5_countries_2022[['Country', '2022 Population']])
```

```python
# Melt the DataFrame to long format for better plotting
df_melted = df.melt(id_vars=['Country', 'Continent'], 
                    value_vars=['1970 Population', '1980 Population', '1990 Population', '2000 Population', '2010 Population',
                                '2015 Population', '2020 Population', '2022 Population'],
                    var_name='Year', value_name='Population')

# Clean the 'Year' column to keep only the year number
df_melted['Year'] = df_melted['Year'].str.extract('(\d{4})').astype(int)

# Filter the melted DataFrame for the top 5 countries
top_5_countries = top_5_countries_2022['Country']
df_top_5 = df_melted[df_melted['Country'].isin(top_5_countries)]

# Plotting the population of the top 5 countries over time
plt.figure(figsize=(12, 8))

# Plot each of the top 5 countries
for country in top_5_countries:
    country_data = df_top_5[df_top_5['Country'] == country]
    plt.plot(country_data['Year'], country_data['Population'], label=country)

plt.title('Population Trends of the Top 5 Most Populous Countries (1970-2022)')
plt.xlabel('Year')
plt.ylabel('Population')
plt.legend()
plt.show()
```

Growth Rates: Calculates the percentage growth from 2000 to 2022 for each country.

```python

# Calculate the growth rate from 2000 to 2022
df['Growth 2000-2022 (%)'] = ((df['2022 Population'] - df['2000 Population']) / df['2000 Population']) * 100

# Display the top 5 rows
print(df[['Country', 'Growth 2000-2022 (%)']].head())
```

Mean Population by Continent: Groups the data by continent and calculates the mean population for 2022.

```python
# Calculate the mean population by continent for 2022
mean_population_continent = df.groupby('Continent')['2022 Population'].mean().reset_index()

# Display the result
print(mean_population_continent)
```

```python
# Calculate the total population by continent for 2022
total_population_continent = df.groupby('Continent')['2022 Population'].sum().reset_index()
total_population_continent = total_population_continent.sort_values(by='2022 Population', ascending=False)

# Display the result
print(mean_population_continent)

```

```python
#find most populated placed and least populated
```

```python
# Find the 10 most populated places in 2022
most_populated = df.nlargest(10, '2022 Population')

# Find the 10 least populated places in 2022
least_populated = df.nsmallest(10, '2022 Population')

# Display the result
print("10 Most Populated Places in 2022:")
print(most_populated[['Country', 'Continent' , '2022 Population']])

# Display the result
print("\n10 Least Populated Places in 2022:")
print(least_populated[['Country', 'Continent', '2022 Population']])
```

correlation coefficient between population density and area.

```python
# Calculate the correlation between density and area
correlation = df['Density (per km²)'].corr(df['Area (km²)'])

# Display the correlation
print(f"Correlation between density and area: {correlation}")
#no strong relation found
```

```python
# Melt the DataFrame to long format
df_melted = df.melt(id_vars=['Country', 'Continent'], 
                    value_vars=['1970 Population', '1980 Population', '1990 Population', '2000 Population',
                                '2010 Population', '2015 Population', '2020 Population', '2022 Population'],
                    var_name='Year', value_name='Population')

# Clean the 'Year' column to keep only the year number
df_melted['Year'] = df_melted['Year'].str.extract(r'(\d{4})').astype(int)
```

```python
# Group by Year and Continent and sum the populations
grouped_df = df_melted.groupby(['Year', 'Continent'])['Population'].sum().reset_index()
```

```python
# Plotting the data
plt.figure(figsize=(14, 8))
sns.lineplot(data=grouped_df, x='Year', y='Population', hue='Continent')

# Add titles and labels
plt.title('Total Population by Continent (1970-2022)')
plt.xlabel('Year')
plt.ylabel('Population (Billion)')
plt.legend(title='Continent')
plt.grid(True)

# Show the plot
plt.show()
```
