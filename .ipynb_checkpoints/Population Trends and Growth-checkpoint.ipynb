```python
# Import Dependencies
import pandas as pd
from pathlib import Path
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the dataset
file_path = Path("Resources/world_population.csv")
df = pd.read_csv(file_path)

df.head()
```

```python
# Calculate the population growth percentage from 1970 to 2022
df['Growth 1970-2022 (%)'] = ((df['2022 Population'] - df['1970 Population']) / df['1970 Population']) * 100
```

```python
# Sort the DataFrame by the growth percentage in descending order
df_sorted = df.sort_values(by='Growth 1970-2022 (%)', ascending=False)

# Select the top 10 countries
top_growth_countries = df_sorted[['Country', 'Continent', '1970 Population', '2022 Population', 'Growth 1970-2022 (%)']].head(10)

# Display the result
print(top_growth_countries)
```

```python
top_growth_countries
```

```python
import matplotlib.pyplot as plt

# Plotting the top 10 countries by population growth
plt.figure(figsize=(12, 8))
plt.bar(top_growth_countries['Country'], top_growth_countries['Growth 1970-2022 (%)'], color='skyblue')
plt.title('Top 10 Countries with Highest Population Growth (1970-2022)')
plt.xlabel('Country')
plt.ylabel('Population Growth (%)')
plt.xticks(rotation=45)
plt.show()
```
