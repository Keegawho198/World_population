```python
world_population_df = pops_df
```

```python
# Calculate growth percentage from 1970 to 2022
# Handling cases with missing data by excluding them from calculations
world_population_df['Growth 1970-2022 (%)'] = ((world_population_df['2022 Population'] - world_population_df['1970 Population']) / 
                                               world_population_df['1970 Population']) * 100

# Rank countries based on the calculated growth percentage, sorting in descending order
growth_rank_df = world_population_df[['Country', '1970 Population', '2022 Population', 'Growth 1970-2022 (%)']].dropna()
growth_rank_df = growth_rank_df.sort_values(by='Growth 1970-2022 (%)', ascending=False).reset_index(drop=True)

# Display the top 10 countries with the greatest growth
growth_rank_df.head(10)
```

```python
# Filter for countries with a negative growth percentage from 1970 to 2022
population_decrease_df = growth_rank_df[growth_rank_df['Growth 1970-2022 (%)'] < 0]

# Display countries with an overall decrease in population
population_decrease_df.reset_index(drop=True)
```

```python
#Check japan
japan_population_data = pops_df


japan_population_data = growth_rank_df[growth_rank_df['Country'] == 'Japan']
japan_population_data
```
